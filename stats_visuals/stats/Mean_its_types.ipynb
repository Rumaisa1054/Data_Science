```python
import pandas as pd
import seaborn as sns
import numpy as np
import sklearn
```

```python
from scipy import stats
arr = np.array([1,2,3,4,5,40])
print(stats.gmean(arr))
print(stats.hmean(arr))
print(stats.trim_mean(arr,proportiontocut=0.1))
```
