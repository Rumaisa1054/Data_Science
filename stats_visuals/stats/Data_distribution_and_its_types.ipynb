# **Why Check the Distribution of Your Data (Data kesa dikhta ha)**

We generally try to make our data normal if its not normal - by removing outliers - cleaning data

Checking the distribution of your data is important because it affects statistical analysis, modeling, and interpretation. Here are the main reasons:

1. **Choice of Statistical Tests**
   - Many tests assume a specific distribution (often normal).  
   - Examples:
     - Parametric tests: t-test, ANOVA, Pearson correlation  assume normality.
     - Non-parametric alternatives: Mann-Whitney U test, Spearman correlation  for non-normal data.
   - If the data is not normal, you may need transformations (log, square root, Box-Cox).

2. **Modeling Assumptions**
   - Many statistical and machine learning models assume certain distributions.
   - Examples:
     - Linear regression  residuals should be approximately normal.
     - Gaussian Naive Bayes  features assumed normal.
     - Time series models  often assume stationarity and specific distribution patterns.

3. **Understanding Data Characteristics**
   - Distribution reveals:
     - Skewness (left or right)
     - Presence of outliers
     - Peaks, flatness, or heavy tails
   - Helps in:
     - Feature engineering
     - Data transformations
     - Improving model accuracy

4. **Parameter Estimation**
   - Certain estimators rely on distribution assumptions to be unbiased or efficient.
   - Example: Standard error formulas assume normality for confidence intervals.

5. **Visualization and Communication**
   - Knowing the distribution helps choose the right visualizations:
     - Histograms, boxplots, density plots
   - Helps decide whether to report mean or median as central tendency.

6. **Data Transformation Decisions**
   - Non-normal distributions may require transformations to meet model assumptions:
     - Log transformation
     - Square root
     - Box-Cox or Yeo-Johnson

7. **Outlier Detection**
   - Skewed or heavy-tailed distributions may indicate outliers.
   - Detecting outliers can improve model performance and robustness.

---

### **Summary**
Checking data distribution is crucial because it:
- Guides method and test selection  
- Ensures model assumptions are met  
- Reveals skewness, outliers, and patterns  
- Improves data interpretation and communication


# **Normal Distribution in Data Science & Data Normalization**

## **Why Normal Distribution is Important**

1. **Foundation for Statistical Tests and Inference**
   - Many parametric tests assume normality:
     - t-test, ANOVA, linear regression, Pearson correlation
   - Confidence intervals and hypothesis testing rely on the **Central Limit Theorem (CLT)**

2. **Central Limit Theorem (CLT)**
   - *The sampling distribution of the mean approaches a normal distribution as sample size increases, regardless of the original distribution.*

3. **Assumption in Machine Learning Models**
   - Some algorithms perform better with normally distributed features:
     - Linear regression, logistic regression, Gaussian Naive Bayes, LDA
   - Models may converge faster and predictions can be more stable

4. **Handling Outliers and Skewed Data**
   - Normalization or transformation reduces skewness and extreme outliers, improving model robustness

5. **Interpretability**
   - Many descriptive statistics (mean, SD, z-scores, confidence intervals) assume approximate normality

---

## **How to Normalize Your Data**

### 1. Min-Max Scaling
- Rescales data to a fixed range, usually [0,1]:
\[
x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}
\]

**Python Example:**
```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

data = np.array([[10], [20], [30]])
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)
print(data_scaled)
````

---

### 2. **Z-Score Standardization**

* Centers data to mean 0 and SD 1:
  
  `z = (x - x_bar) / s`
  

**Python Example:**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
print(data_scaled)
```

---

### 3. Log Transformation

* Reduces right skewness:
  
  `x' = log(x + 1)`

**Python Example:**

```python
import numpy as np

data = np.array([1, 10, 100])
data_log = np.log1p(data)
print(data_log)
```

---

### 4. Box-Cox Transformation

* Finds optimal power transformation to make data more normal
* Only works with positive values

**Python Example:**

```python
from scipy.stats import boxcox

data = np.array([1, 2, 3, 4, 5])
data_boxcox, _ = boxcox(data)
print(data_boxcox)
```

---

### 5. Yeo-Johnson Transformation

* Similar to Box-Cox but works with zero or negative values

**Python Example:**

```python
from sklearn.preprocessing import PowerTransformer

data = np.array([[0], [1], [-1], [2]])
pt = PowerTransformer(method='yeo-johnson')
data_transformed = pt.fit_transform(data)
print(data_transformed)
```

---

## **Quick Summary Table**

| Method                  | Formula / Concept                                  | Use Case                                                      |
| ----------------------- | -------------------------------------------------- | ------------------------------------------------------------- |
| Min-Max Scaling         | (x_{norm} = x - x_min / x_max - x_min | Scale features to [0,1], useful for distance-based algorithms |
| Z-Score Standardization | (z = x - x_bar / s)                        | Center data, assume normality, compare features               |
| Log Transformation      | x' = log(x + 1)                                 | Reduce right skewness, handle positive skewed data            |
| Box-Cox                 | Optimal power transform                            | Reduce skewness, approximate normality, positive data only    |
| Yeo-Johnson             | Power transform including zero/negative            | Reduce skewness for general data                              |



# **Rule of Thumb:**

If a model/test assumes normality, transform skewed continuous data.

If a model/test is designed for the data type (count, fraction, binary, positive skew), keep the original distribution.

# **Statistical Tests and Models Assuming Normality**

## **1. Basic Statistical Tests**

| Test                                | Purpose / Notes                                             |
| ----------------------------------- | ----------------------------------------------------------- |
| **One-sample t-test**               | Compare sample mean to a known population mean              |
| **Two-sample t-test (independent)** | Compare means of two independent groups                     |
| **Paired t-test (dependent)**       | Compare means of two related groups                         |
| **ANOVA (one-way, multi-way)**      | Compare means across 3+ groups                              |
| **Z-test**                          | Compare sample mean to population mean when variance known  |
| **Pearson correlation**             | Measure linear correlation between two continuous variables |
| **Linear contrast tests**           | Compare specific linear combinations of group means         |

---

## **2. Regression Models**

| Model                                    | Assumptions / Notes                                           |
| ---------------------------------------- | ------------------------------------------------------------- |
| **Linear Regression**                    | Residuals should be approximately normally distributed        |
| **Multiple Linear Regression**           | Same as linear regression, with multiple predictors           |
| **Gaussian Naive Bayes**                 | Assumes each feature is normally distributed                  |
| **LDA (Linear Discriminant Analysis)**   | Assumes predictors are normally distributed within each class |
| **Ridge / Lasso Regression (OLS-based)** | Assumes residuals are normal for inference (CI, p-values)     |

---

## **3. Confidence Intervals & Effect Size**

| Method                             | Notes                                             |
| ---------------------------------- | ------------------------------------------------- |
| **Confidence intervals for means** | SE formulas assume normality (or large n for CLT) |
| **Cohens d / effect sizes**       | Normality assumption improves interpretability    |

---

## **4. Other Tests / Multivariate**

| Test / Model                                  | Notes                                               |
| --------------------------------------------- | --------------------------------------------------- |
| **MANOVA (Multivariate ANOVA)**               | Multivariate normality required for predictors      |
| **Hotellings T test**                       | Multivariate normality assumption                   |
| **Regression diagnostics / hypothesis tests** | t-tests on coefficients assume residuals are normal |

---

## **5. Key Notes**

1. **Central Limit Theorem (CLT)** often allows approximate normality if **sample size is large** (n > 30-50), even if raw data isnt normal.
2. If data is **skewed**, use transformations: **log, square root, Box-Cox, Yeo-Johnson**.
3. Many machine learning models (tree-based methods, random forest, XGBoost) **do not require normality**.

---

**Summary:**
**Tests and models that expect normality mainly include:**

* **t-tests, ANOVA, Z-tests, Pearson correlation**
* **Linear regression and OLS-based models**
* **LDA, Gaussian Naive Bayes**
* **Confidence intervals and multivariate normal tests**


# **tests and models that do NOT assume normality**
Often called **non-parametric tests** or **distribution-free methods**. These are important when your data is **skewed, ordinal, or discrete**.

Heres a comprehensive list:

---

# **1. Non-Parametric / Distribution-Free Tests**

| Test                                | Purpose / Notes                                                                           |
| ----------------------------------- | ----------------------------------------------------------------------------------------- |
| **Mann-Whitney U test**             | Compare two independent groups (alternative to independent t-test)                        |
| **Wilcoxon signed-rank test**       | Compare two related groups (alternative to paired t-test)                                 |
| **Kruskal-Wallis test**             | Compare 3+ independent groups (alternative to ANOVA)                                      |
| **Friedman test**                   | Compare 3+ related groups (alternative to repeated measures ANOVA)                        |
| **Spearman rank correlation**       | Measure monotonic relationship between two variables (alternative to Pearson correlation) |
| **Kendalls tau**                   | Measure rank correlation between variables                                                |
| **Sign test**                       | Test median of a sample or difference between paired samples                              |
| **`Chi-Square test of independence`** | Test association between categorical variables                                            |
| ` **`Chi-Square goodness-of-fit`** `      | Compare observed vs expected frequencies                                                  |
| **Fishers exact test**             | Test association in small categorical datasets                                            |

---

# **2. Regression / Modeling Without Normality Assumptions**

| Model                               | Data Type                        | Notes                                                             |
| ----------------------------------- | -------------------------------- | ----------------------------------------------------------------- |
| **`Logistic Regression`**             | Binary outcome                   | Works with 0-1 outcomes, no normality required                    |
| **Poisson Regression**              | Count outcome                    | For count data; assumes Poisson distribution                      |
| **Negative Binomial Regression**    | Count data with overdispersion   | Alternative to Poisson regression                                 |
| **Beta Regression**                 | Fraction / proportion data [0,1] | Models probabilities directly                                     |
| **Quantile Regression**             | Continuous data                  | Models conditional quantiles, robust to skew/outliers             |
| **Generalized Linear Models (GLM)** | Various outcome distributions    | Use link functions appropriate for outcome (log, logit, identity) |

---

# **3. Other Non-Parametric / Robust Methods**

| Method                                    | Notes                                                 |
| ----------------------------------------- | ----------------------------------------------------- |
| **`Random Forest / Decision Trees`**        | No distributional assumptions                         |
| **`Gradient Boosting (XGBoost, LightGBM)`** | Works with skewed or categorical data                 |
| **`K-Nearest Neighbors`**                   | Non-parametric; depends on distance, not distribution |
| **`Bootstrap / Resampling Methods`**        | Estimate SE, CI without normality assumption          |

---

# **4. Key Notes**

1. Use these methods when your data is **non-normal, ordinal, skewed, or discrete**.
2. Non-parametric tests are **less powerful** if the data actually is normal, but they are **safer for skewed or unknown distributions**.
3. Many **modern ML models** do not require normality at all.

---

**Summary Table**

| Data Type / Distribution | Use Non-Normal Methods? | Example Tests / Models                               |
| ------------------------ | ----------------------- | ---------------------------------------------------- |
| Skewed continuous        | Yes                     | Mann-Whitney, Kruskal-Wallis, Quantile regression    |
| Paired non-normal        | Yes                     | Wilcoxon signed-rank, Sign test                      |
| Counts                   | Yes                     | Poisson regression, Negative Binomial regression     |
| Binary / proportions     | Yes                     | Logistic regression, Beta regression                 |
| Categorical              | Yes                     | Chi-Square test, Fishers exact test                 |
| Ordinal                  | Yes                     | Spearman / Kendall correlation, non-parametric ANOVA |
| Any complex pattern      | Yes                     | Random forest, gradient boosting, KNN, bootstrap     |


# **Tests/Models That Require Normality**
- One-sample t-test
- Two-sample t-test
- Paired t-test
- ANOVA (all types)
- Pearson correlation
- Z-test
- ANCOVA
- Linear Regression (requires normality of residuals, not predictors)

# **Tests/Models That Do NOT Require Normality**
- Mann-Whitney U
- Wilcoxon Signed-Rank
- Kruskal-Wallis
- Friedman Test
- Spearman/Kendall correlation
- Chi-Square, Fisher Exact

# **Machine Learning Models (No Normality Required)**
- Logistic Regression
- Poisson/Gamma/Beta/Negative Binomial (GLMs)
- Random Forest, Decision Trees
- XGBoost / LightGBM
- SVM
- KNN
- Neural Networks
- Clustering (KMeans, DBSCAN)

# **Key Point**
Only classical **parametric statistical tests** need normality.  
Machine learning models do **not** require normality.
