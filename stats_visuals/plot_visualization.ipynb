```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/observablehq/sample-datasets/main/diamonds.csv"
df = pd.read_csv(url)

df.head()
```

```python
plt.plot(df["carat"],df["price"])
plt.show()
```

# **xlabel and ylabel**

```python
plt.plot(df["carat"],df["price"])
plt.xlabel("carat")
plt.ylabel("price")
plt.show()
```

# **title**

```python
plt.plot(df["carat"],df["price"])
plt.title("carat vs price")
plt.show()
```

# **legend**

```python
sns.scatterplot(data=df,x="carat",y="price",hue='cut')
plt.legend()
plt.show()
```

```python
sns.scatterplot(data=df,x="carat",y="price",hue='cut')
plt.legend(title = "cut-type",loc='lower right')
plt.show()
```

# **Ticks**

```python
sns.scatterplot(data=df,x="carat",y="price",hue='cut')
plt.legend(title = "cut-type",loc='lower right')
plt.xticks(rotation=45,color='blue')
plt.show()
```

```python
sns.scatterplot(data=df,x="carat",y="price",hue='cut')
plt.legend(title = "cut-type",loc='lower right')
plt.xticks(rotation=45,color='blue',fontsize=15)
plt.yticks(color='red')
plt.show()
```

# **grid**

```python
sns.scatterplot(data=df,x="carat",y="price",hue='cut')
plt.grid(True,color="red",alpha=0.5)
plt.show()
```

```python
sns.scatterplot(data=df,x="carat",y="price",hue='cut')
plt.grid(True,color="red",alpha=0.5,linestyle="--")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x="carat",y="price",hue='cut')
plt.grid(True,color="red",alpha=0.5,linestyle="--")
plt.show()
```

# **palette**

```python
sns.set_palette("deep") # muted, deep,pastel,bright,dark,colorbind
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x="carat",y="price",hue='cut')
plt.grid(True,color="red",alpha=0.5,linestyle="--")
plt.show()
```
